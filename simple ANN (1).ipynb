```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

import dataset

```python
dataset = pd.read_csv('/content/drive/MyDrive/datasetsss/Churn_Modelling.csv')
```

```python
X= dataset.iloc[:, 3:13]
y= dataset.iloc[:, 13]
```

create dummy variables

```python
geography= pd.get_dummies(X["Geography"], drop_first= True)
gender= pd.get_dummies(X['Gender'], drop_first= True)
```

Concatenate the data frames

```python
X= pd.concat([X, geography, gender], axis=1)
```

drop unnessecary coloumns

```python
X= X.drop(['Geography', 'Gender'], axis= 1)
```

splitting the dataset into training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 0)
```

feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

importing the keras libaries and packages

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

intializing the ANN model

```python
classi = Sequential()

classi.add(Dense(units= 64, kernel_initializer= 'he_normal', activation= 'relu', input_dim= 11))
classi.add(Dropout(0.2))
classi.add(Dense(units= 32, kernel_initializer= 'he_normal', activation= 'relu'))
classi.add(Dropout(0.1))
classi.add(Dense(units= 16, kernel_initializer= 'he_normal', activation= 'relu'))
classi.add(Dropout(0.2))
classi.add(Dense(units= 8, kernel_initializer= 'he_normal', activation= 'relu'))
classi.add(Dropout(0.2))
classi.add(Dense(units= 4, kernel_initializer= 'he_normal', activation= 'relu'))
classi.add(Dropout(0.2))
classi.add(Dense(units= 1, kernel_initializer= 'glorot_uniform', activation= 'sigmoid'))
```

compile the model

```python
classi.compile(optimizer='adam', loss='binary_crossentropy', metrics= 'accuracy')
```

summary of the model

```python
classi.summary()
```

fitting the model

```python
history = classi.fit(X_train, y_train, epochs=100, validation_split= 0.33 , batch_size=10)
```

```python
y_pred= classi.predict(X_test)
y_pred= (y_pred >0.5)

from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, y_pred)
print(cm)

from sklearn.metrics import accuracy_score
score = accuracy_score(y_pred, y_test)
print(score)
```
